```python
class SC(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self,X,y=None):
        self.means = X.mean()
        self.stds = X.std(ddof=0)
        return self
    
    def transform(self, X, y=None):
        return (X-self.means)/self.stds
```

```python
sc = SC()
sc.fit_transform(X_train)
sc.transform(X_test)
sc.means
```
